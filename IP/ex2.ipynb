```python
from collections import Counter

documents = [
    "natural language processing is interesting",
    "machine learning and natural language processing",
    "deep learning for language understanding"
]

query = "language processing"

query_words = query.split()

scores = []

for doc in documents:
    words = doc.split()
    count = Counter(words)
    score = 0
    
    for word in query_words:
        if word in count:
            score = score + count[word]
    
    scores.append(score)

print("TF Ranking:")
for i in range(len(documents)):
    print("Score:", scores[i], "Document:", documents[i])
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer

documents = [
    "natural language processing is interesting",
    "machine learning and natural language processing",
    "deep learning for language understanding"
]

query = "language processing"

vectorizer = TfidfVectorizer()

tfidf_matrix = vectorizer.fit_transform(documents + [query])

query_vector = tfidf_matrix[-1]
doc_vectors = tfidf_matrix[:-1]

scores = []

for i in range(doc_vectors.shape[0]):
    score = doc_vectors[i].dot(query_vector.T).toarray()[0][0]
    scores.append(score)

print("TF-IDF Ranking")
for i in range(len(documents)):
    print("Score:", scores[i], "Document:", documents[i])
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.metrics.pairwise import cosine_similarity

documents = [
    "natural language processing is interesting",
    "machine learning and natural language processing",
    "deep learning for language understanding"
]

query = "language processing"

vectorizer = TfidfVectorizer()

tfidf_matrix = vectorizer.fit_transform(documents + [query])

doc_vectors = tfidf_matrix[:-1]
query_vector = tfidf_matrix[-1]

similarity = cosine_similarity(doc_vectors, query_vector)

print("Cosine Similarity Ranking")
for i in range(len(documents)):
    print("Score:", similarity[i][0], "Document:", documents[i])
```
