```python
import nltk
import string
from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize
from collections import Counter

# Download required resources (run once)
nltk.download('punkt')
nltk.download('stopwords')

text = """Natural language processing is a field of artificial intelligence
that focuses on interaction between computers and human language."""

# 1. Word Tokenization
tokens = word_tokenize(text)

# 2. Punctuation Removal
tokens_no_punct = [word for word in tokens if word not in string.punctuation]

# 3. Stopword Removal
stop_words = set(stopwords.words('english'))
filtered_words = [word for word in tokens_no_punct if word.lower() not in stop_words]

# 4. Topical Word Extraction (Most frequent words)
word_freq = Counter(filtered_words)
top_words = word_freq.most_common(5)

print("Tokens:", tokens)
print("After removing punctuation:", tokens_no_punct)
print("After removing stopwords:", filtered_words)
print("Topical words:", top_words)
```

```python
import string
from torchnlp.encoders.text import WhitespaceEncoder
from nltk.tokenize import word_tokenize
from nltk.corpus import stopwords
from collections import Counter
import nltk

nltk.download('punkt')
nltk.download('stopwords')

text = """Natural language processing helps computers understand human language
and process text efficiently."""

# Word Tokenization
tokens = word_tokenize(text)

# Remove punctuation
tokens = [word for word in tokens if word not in string.punctuation]

# Remove stopwords
stop_words = set(stopwords.words('english'))
tokens = [word for word in tokens if word.lower() not in stop_words]

# Encoding words using PyTorch-NLP
encoder = WhitespaceEncoder(tokens)
encoded = [encoder.encode(word) for word in tokens]

# Topical words using frequency
freq = Counter(tokens)
top_words = freq.most_common(5)

print("Processed Tokens:", tokens)
print("Encoded Words:", encoded)
print("Topical Words:", top_words)
```
