```python
# Advanced Rule-Based Chatbot with Context Handling

import re

def chatbot():
    print("Chatbot: Hello! I am an advanced rule-based chatbot.")
    print("Chatbot: You can ask about services, timings, or contact.")
    print("Chatbot: Type 'bye' to exit.\n")

    context = None  # Stores conversation state

    while True:
        user_input = input("You: ").lower()

        if re.search(r"\b(bye|exit|quit)\b", user_input):
            print("Chatbot: Goodbye! Have a great day.")
            break

        # Greeting intent
        elif re.search(r"\b(hi|hello|hey)\b", user_input):
            print("Chatbot: Hello! How can I help you today?")
            context = "greeted"

        # Services intent
        elif re.search(r"\b(services|service)\b", user_input):
            print("Chatbot: We offer support, training, and consultation.")
            print("Chatbot: Would you like details on any service?")
            context = "services"

        # Context-based response
        elif context == "services" and re.search(r"\b(training)\b", user_input):
            print("Chatbot: Our training programs include Python and NLP.")
            context = None

        elif context == "services" and re.search(r"\b(support)\b", user_input):
            print("Chatbot: We provide 24/7 technical support.")
            context = None

        # Timing intent
        elif re.search(r"\b(time|timing|hours)\b", user_input):
            print("Chatbot: We are open from 9 AM to 6 PM, Monday to Friday.")

        # Contact intent
        elif re.search(r"\b(contact|phone|email)\b", user_input):
            print("Chatbot: You can contact us at support@example.com.")

        # Help intent
        elif re.search(r"\b(help|options|menu)\b", user_input):
            print("Chatbot: You can ask about services, timings, or contact info.")

        # Fallback
        else:
            print("Chatbot: I'm sorry, I didn't understand that.")
            print("Chatbot: Please try asking about services, timings, or contact.")

# Run chatbot
chatbot()
```

```python
# Simple Rule-Based Chatbot

def chatbot():
    print("Chatbot: Hello! I am a simple chatbot.")
    print("Chatbot: Type 'bye' to exit.\n")

    while True:
        user_input = input("You: ").lower()

        if user_input in ["hi", "hello", "hey"]:
            print("Chatbot: Hello! How can I help you?")

        elif "name" in user_input:
            print("Chatbot: I am a simple rule-based chatbot.")

        elif "help" in user_input:
            print("Chatbot: I can answer basic questions. Try asking about my name.")

        elif user_input in ["bye", "exit", "quit"]:
            print("Chatbot: Goodbye! Have a nice day.")
            break

        else:
            print("Chatbot: Sorry, I did not understand that.")

# Run the chatbot
chatbot()
```
