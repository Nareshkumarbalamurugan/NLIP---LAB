```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Image path
image_path = '/content/industrial-machine-260nw-47955199.webp'

# Read image in grayscale
img = cv2.imread(image_path, cv2.IMREAD_COLOR)

# Display original image
plt.imshow(img, cmap='gray')
plt.title("Original Image")
plt.axis('off')
```

```python
# Low Pass Filter kernel
lpf = np.ones((3,3), np.float32) / 9

# Apply filter
lpf_img = cv2.filter2D(img, -1, lpf)

# Display result
plt.imshow(lpf_img, cmap='gray')
plt.title("Low Pass Filtered Image")
plt.axis('off')
```

```python
# High Pass Filter kernel
hpf = np.array([[-1, -1, -1],
                [-1,  8, -1],
                [-1, -1, -1]])

# Apply filter
hpf_img = cv2.filter2D(img, -1, hpf)

# Display result
plt.imshow(hpf_img, cmap='gray')
plt.title("High Pass Filtered Image")
plt.axis('off')
```

```python
plt.figure(figsize=(12,4))

plt.subplot(1,3,1)
plt.imshow(img, cmap='gray')
plt.title("Original")
plt.axis('off')

plt.subplot(1,3,2)
plt.imshow(lpf_img, cmap='gray')
plt.title("Low Pass Filter")
plt.axis('off')

plt.subplot(1,3,3)
plt.imshow(hpf_img, cmap='gray')
plt.title("High Pass Filter")
plt.axis('off')

plt.show()
```

```python
# High Pass Filter
hpf_img = cv2.filter2D(img, cv2.CV_64F, hpf)

# Sharpened image
sharpened = img + hpf_img

# Normalize
sharpened = cv2.normalize(sharpened, None, 0, 255, cv2.NORM_MINMAX)
sharpened = np.uint8(sharpened)

# Display
plt.imshow(sharpened, cmap='gray')
plt.title("Sharpened Bone Image")
plt.axis('off')
```
