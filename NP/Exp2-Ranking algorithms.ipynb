```python
from collections import Counter
import re

# Sample text
text = "Ranking algorithms rank items based on importance ranking improves search results"

# Preprocessing
words = re.findall(r'\b\w+\b', text.lower())

# Frequency count
frequency = Counter(words)

# Ranking words by frequency
ranked_words = frequency.most_common()

print("Frequency-Based Ranking:")
for word, count in ranked_words:
    print(f"{word} -> {count}")
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer

# Sample documents
documents = [
    "ranking algorithms are important",
    "ranking improves search engines",
    "algorithms rank documents"
]

# TF-IDF Vectorization
vectorizer = TfidfVectorizer()
tfidf_matrix = vectorizer.fit_transform(documents)

# Feature names and scores
feature_names = vectorizer.get_feature_names_out()
scores = tfidf_matrix.sum(axis=0).A1

# Ranking terms by TF-IDF score
tfidf_ranking = sorted(zip(feature_names, scores), key=lambda x: x[1], reverse=True)

print("\nTF-IDF Ranking:")
for word, score in tfidf_ranking:
    print(f"{word} -> {score:.3f}")
```

```python
import networkx as nx

# Create graph
graph = nx.DiGraph()

# Add edges (links)
graph.add_edges_from([
    ('A', 'B'),
    ('B', 'C'),
    ('C', 'A'),
    ('A', 'C')
])

# Apply PageRank
pagerank_scores = nx.pagerank(graph)

print("\nPageRank Scores:")
for node, score in pagerank_scores.items():
    print(f"{node} -> {score:.3f}")
```
