```python
from nltk.corpus import stopwords

stop_words = set(stopwords.words('english'))

filtered_words = [word for word in tokens_no_punct if word.lower() not in stop_words]
print(filtered_words)
```

```python
import string
from torchnlp.encoders.text import WhitespaceEncoder
from collections import Counter

# Sample text
text = "Natural Language Processing helps computers understand human language."

# Lowercase
text = text.lower()

# Punctuation removal
text = text.translate(str.maketrans('', '', string.punctuation))

# Tokenization
tokens = text.split()

# Stop-word removal
stop_words = {'the', 'is', 'and', 'to', 'in', 'of', 'a', 'helps'}
filtered_tokens = [word for word in tokens if word not in stop_words]

# Convert tokens back to string
filtered_text = " "                               .join(filtered_tokens)

# PyTorch-NLP Encoding
encoder = WhitespaceEncoder(filtered_text.split())
encoded_tokens = encoder.encode(filtered_text)

# Topical word extraction
word_freq = Counter(filtered_tokens)
topical_words = word_freq.most_common(5)

print("Filtered Tokens:", filtered_tokens)
print("Encoded Tokens:", encoded_tokens)
print("Topical Words:", topical_words)
```

```python
text = """
Natural Language Processing (NLP) is a subfield of Artificial Intelligence.
It focuses on the interaction between computers and human language!
"""
```

```python
def remove_punctuation(text):
  punctuation = ".,!?;:'\"()[]{}<>-_/\\"
  cleaned_text=""
  for char in text:
    if char not in punctuation:
      cleaned_text+=char
  return cleaned_text
text_no_punct = remove_punctuation(text)
print(text_no_punct)
```

```python
text_lower=text_no_punct.lower()
print(text_lower)
```

```python
def manual_tokenization(text):
  tokens=[]
  word=""
  for char in text:
    if char==" " or char=="\n":
      if word!="":
        tokens.append(word)
        word=""
    else:
      word+=char
  if word!="":
    tokens.append(word)
  return tokens
tokens=manual_tokenization(text_lower)
print(tokens)
```

```python
stop_words = [
    "is", "a", "of", "the", "on", "and", "it",
    "between", "to", "in", "for", "with"
]
```

```python
def remove_stopwords(tokens, stop_words):
    filtered_tokens = []

    for word in tokens:
        if word not in stop_words:
            filtered_tokens.append(word)

    return filtered_tokens

filtered_tokens = remove_stopwords(tokens, stop_words)
print(filtered_tokens)
```

```python
def calculate_frequency(tokens):
  freq={}
  for word in tokens:
    if word in freq:
      freq[word]+=1
    if word not in freq:
      freq[word]=1
  return freq
frequency = calculate_frequency(filtered_tokens)
print(frequency)
```
